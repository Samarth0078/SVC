```python
cols=['Gender','Married','Dependents','Education','Self_Employed','Loan_Amount_Term','Credit_History']
for col in cols:
  df[col]=df[col].fillna(df[col].mode()[0])
df
```

```python
df['Self_Employed']=df['Self_Employed'].map({'No':0,'Yes':1})
df
```

```python
df['Property_Area'].unique()
```

```python
df['Property_Area']=df['Property_Area'].map({'Urban':1,'Rural':0,'Semiurban':2})
df
```

```python
# 1.Read Dataset
# 2.Check is there any missing values
# 3.Ensure all the columns having numeric datatype
# 4.Split the data into input(features) and output(target)
# 5.Scaling of data
# 6.Split the datset into training and testing dataset
# 7.Create model using an algorithm and train data
# 8.Check the perfomance of the model
# 9.Predict with new values
```

```python
X=df.iloc[:,1:-1]
X
```

```python
y=df.iloc[:,-1]
y
```

```python
#standardScaler
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
X_scaled
```

```python
from numpy.random.mtrand import RandomState
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.3,random_state=1)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(X_train,y_train)
```

```python
y_pred=knn.predict(X_test)
y_pred
```

```python
y_test.values
```

```python
from sklearn.metrics import accuracy_score,ConfusionMatrixDisplay
print("Accuracy :",accuracy_score(y_test,y_pred)*100)
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred,digits=3))
```

```python
#SVC
from sklearn.svm import SVC
sv=SVC()
sv.fit(X_train,y_train)
```

```python
y_pred=sv.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report,ConfusionMatrixDisplay
print(accuracy_score(y_test,y_pred)*100)
print(classification_report(y_test,y_pred,digits=3))
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
```
